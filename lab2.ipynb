```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import torch
import torchvision
import torchvision.transforms.functional as TF
```

```python
def load_image(image_path):
    image = torchvision.io.read_image(image_path)
    return image
```

## Affine Transformation

```python
image =load_image("pabin.jpeg")

rotated_image = TF.rotate(image,  angle=45, expand=False)
reflected_image = TF.hflip(image)
resized_image = TF.resize(image, [200, 200])
scaled_image = TF.affine(image, angle=0, translate=[0, 0], scale=2, shear=[0, 0])
sheared_image = TF.affine(image, angle=0, translate=[0, 0], scale=1, shear=[10, 0])
```

```python
plt.subplot(2,2,1)
plt.title('Original Image')
plt.imshow(image.permute(1, 2, 0))
plt.axis('off')  

plt.subplot(2, 2, 2)
plt.title('Rotated Image')
plt.imshow(rotated_image.permute(1, 2, 0))
plt.axis('off')  

plt.subplot(2, 2, 3)
plt.title('Scaled Image')
plt.imshow(scaled_image.permute(1, 2, 0))
plt.axis('off')  

plt.subplot(2, 2, 4)
plt.title('Sheared Image')
plt.imshow(sheared_image.permute(1, 2, 0))
plt.axis('off')  # Hide axes
plt.show()
```

```python
plt.imshow(resized_image.permute(1, 2, 0))
plt.title('Resized Image')
plt.axis('off')  # Hide axes
plt.show()
```

## Color

```python
b_image = TF.adjust_brightness(image, brightness_factor=0.5)
c_image = TF.adjust_contrast(image, contrast_factor=2.0)
s_image = TF.adjust_saturation(image, saturation_factor=2.0)
h_image = TF.adjust_hue(image, hue_factor=0.1)
```

```python
plt.subplot(2,2,1)
plt.title('Brightness Adjusted Image')
plt.imshow(b_image.permute(1, 2, 0))
plt.axis('off')  

plt.subplot(2, 2, 2)
plt.title('Contrast Adjusted Image')
plt.imshow(c_image.permute(1, 2, 0))
plt.axis('off')  

plt.subplot(2, 2, 3)
plt.title('Saturation Adjusted Image')
plt.imshow(s_image.permute(1, 2, 0))
plt.axis('off')  

plt.subplot(2, 2, 4)
plt.title('Hue Adjusted Image')
plt.imshow(h_image.permute(1, 2, 0))
plt.axis('off')  
plt.show()
```
